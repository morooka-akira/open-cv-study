```python
import cv2 as cv
import matplotlib.pyplot as plt

matplotlib inline
```

```python
def show_img(src):
    # opencvの画像はBGR形式なのでRGBへ変換して表示する
    plt.imshow(cv.cvtColor(src, cv.COLOR_BGR2RGB))
    plt.show()
```

```python
image_file = 'files/target.jpg'
# 画像の読み込み
src = cv.imread(image_file, cv.IMREAD_COLOR)
# 元画像の表示
show_img(src)
```

```python
# グレースケール
dst = cv.cvtColor(src, cv.COLOR_RGB2GRAY)
show_img(dst)
```

```python
# 単純なしきい値フィルタリング
# THRESH_BINARY：　閾値を超えるピクセルは maxVal に，それ以外のピクセルは 0 になります．
# THRESH_BINARY_INV：　閾値を超えるピクセルは 0 に，それ以外のピクセルは maxVal になります．
# THRESH_TRUNC:　閾値を超えるピクセルは threshold に，それ以外のピクセルは変更されません．
# THRESH_TOZERO：　閾値を超えるピクセルは変更されず，それ以外のピクセルは 0 になります．
# THRESH_TOZERO_INV：　閾値を超えるピクセルは 0 に，それ以外のピクセルは変更されません．
retval, dst = cv.threshold(dst, 127, 255, cv.THRESH_TOZERO)
show_img(dst)
```

```python
 # 白黒の反転
dst = cv.bitwise_not(dst)
show_img(dst)
```

```python
# 自動でしきい値を決めてフィルタリング(THRESH_OTSU:大津の二値化処理)を使う
retval, dst = cv.threshold(dst, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
print("ret: {}".format(retval))
show_img(dst)
```
